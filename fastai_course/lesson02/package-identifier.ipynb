```python
#/export
from fastai.vision.all import *
import gradio as gr

```

```python
im = PILImage.create('pepsi_can.jpeg')
im.thumbnail((192,192))
im
```

```python
#/export
learn = load_learner('model.pkl')
```

```python
learn.predict(im)
```

```python
#/export
categories = ('Can', 'Pet')

def classify_image(img):
    pred,idx, probs = learn.predict(img)
    return dict(zip(categories, map(float, probs)))
```

```python
classify_image(im)
```

```python
#/export
can = PILImage.create('pepsi_can.jpeg')
can.thumbnail((192,192))

pet = PILImage.create('pet_bottle.jpg')
pet.thumbnail((192,192))

examples = [pet, can]

intf = gr.Interface(fn=classify_image, inputs="image", outputs="label", examples=examples)
intf.launch(inline=False)
```
